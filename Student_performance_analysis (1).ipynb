```python
labels = ['Male','Female']
sizes = [data.query('gender == "male"').gender.count(),data.query('gender == "female"').gender.count()]
#colors
colors = ['#ffdaB9','#66b3ff']
#explsion
explode = (0.05,0.05)
plt.figure(figsize=(8,8)) 
my_circle=plt.Circle( (0,0), 0.7, color='white')
plt.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85,explode=explode)
p=plt.gcf()
plt.axis('equal')
p.gca().add_artist(my_circle)
plt.show()
```

```python
def power_train(dataframe): 
    sns.countplot(x= dataframe['test preparation course'])
    plt.title('test preparation course', fontsize = 20)
    plt.xlabel('completed or not', fontsize = 15)
    plt.ylabel('Count', fontsize = 15)
```

```python
def bodystyle(dataframe):
    plt.figure(figsize=(10, 5))
    sns.countplot(x= 'race/ethnicity', data= dataframe, hue='test preparation course')
    plt.title('Count plot of course completed according to group', fontsize= 20)
    plt.xlabel('group', fontsize= 15)
    plt.ylabel('Count', fontsize= 15)
    plt.show()
```

```python
# Count Plots
power_train(data)
bodystyle(data)
```

```python
def race_ethnicity(dataframe): 
    sns.countplot(x= dataframe['race/ethnicity'])
    plt.title('race_ethnicity', fontsize = 20)
    plt.xlabel('Group', fontsize = 15)
    plt.ylabel('Count', fontsize = 15)
```

```python
def marks(dataframe):
    plt.figure(figsize=(15, 5))
    sns.countplot(x= ('race/ethnicity') , data= dataframe, hue='math score')
    plt.title('Count plot of marks obtained in math score according to group', fontsize= 20)
    plt.xlabel('group', fontsize= 15)
    plt.ylabel('Count', fontsize= 15)
    plt.show()
```

```python
race_ethnicity(data)
marks(data)
```
